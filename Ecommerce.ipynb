```python
import pandas as pd

df = pd.read_excel('ecommerce_data.csv.xlsx')
print(df.head())
print(df.shape)
```

```python
import matplotlib.pyplot as plt

# Chart 1 - Revenue by Category
cat_revenue = df.groupby('category')['total_price'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
cat_revenue.plot(kind='bar', color='steelblue')
plt.title('Revenue by Category', fontsize=14)
plt.xlabel('Category')
plt.ylabel('Total Revenue (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('revenue_by_category.png', dpi=150)
plt.show()

# Chart 2 - Revenue by City
city_revenue = df.groupby('city')['total_price'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
city_revenue.plot(kind='bar', color='coral')
plt.title('Revenue by City', fontsize=14)
plt.xlabel('City')
plt.ylabel('Total Revenue (₹)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('revenue_by_city.png', dpi=150)
plt.show()

# Chart 3 - Payment Methods
payment = df.groupby('payment_method')['order_id'].count().sort_values(ascending=False)

plt.figure(figsize=(8, 8))
payment.plot(kind='pie', autopct='%1.1f%%')
plt.title('Payment Method Distribution', fontsize=14)
plt.tight_layout()
plt.savefig('payment_methods.png', dpi=150)
plt.show()

print("All charts saved!")
```
